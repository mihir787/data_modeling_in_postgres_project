```python
import os
import pandas as pd
from datetime import datetime
```

```python
# get song data files

files_to_ignore = [".DS_Store"]

log_files = []
log_df = pd.DataFrame()
for r, d, f in os.walk('data/log_data'):
    for file in f:
        file_path = os.path.join(r, file)
        filename = file_path.split('/')[-1]
        if filename not in files_to_ignore:
            temp_data = pd.read_json(file_path, lines = True)
            log_df = log_df.append(temp_data, ignore_index = True)

song_files = []
song_df = pd.DataFrame()
for r, d, f in os.walk('data/song_data'):
    for file in f:
        file_path = os.path.join(r, file)
        filename = file_path.split('/')[-1]
        if filename not in files_to_ignore:
            temp_data = pd.read_json(file_path, lines = True)
            song_df = song_df.append(temp_data, ignore_index = True)
```
